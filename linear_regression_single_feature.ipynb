# Linear regression model 2

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('seaborn')
```

```python
ds=pd.read_csv("/content/train.csv")
ds
```

```python
ds.describe()
```

```python
print(ds['y'].shape)
```

```python
print(ds)
```

```python
theta0=0
theta1=0
def hypothesis(x1,theta0,theta1):
  return (theta0 + theta1*x1)
```

```python

plt.scatter(ds['x'],ds['y'])
plt.plot(ds['x'],hypothesis(ds['x'],theta0,theta1))
plt.xlim(-5,105)
```

```python
def gradient_theta0(theta0,theta1):
  sum1=0
  for i in range(len(ds['x'])):
    sum1+=(hypothesis(ds['x'][i],theta0,theta1)-ds['y'][i])
  return(sum1 / 700)
```

```python
gradient_theta0(0,0)
```

```python
def gradient_theta1(theta0,theta1):
  sum2=0
  for i in range(len(ds['x'])):
    sum2+=(hypothesis(ds['x'][i],theta0,theta1) - ds['y'][i]) * ds['x'][i]
  return (sum2 / 700)
```

```python
gradient_theta1(0,0)
```

```python
def linear(learning_rate):
  theta0_temp=0
  theta1_temp=0
  theta0=0
  theta1=0
  t1=[]
  t0=[]
  plt.scatter(ds['x'],ds['y'])
  for i in range(300):
    theta0_temp=theta0 - learning_rate * (gradient_theta0(theta0,theta1))
    theta1_temp=theta1 - learning_rate * (gradient_theta1(theta0,theta1))
    print(theta0_temp,theta1_temp)
    theta1=theta1_temp
    theta0=theta0_temp
    t1.append(theta1)
    t0.append(theta0)
    plt.plot(ds['x'],hypothesis(ds['x'],theta0,theta1))
  print(t1,t0)
  return theta0,theta1

```

```python
theta0_list,theta1_list=linear(0.0001)
```

```python
print(theta0_list,theta1_list)
```

```python
theta0_detail,theta1_detail=linear(0.00001)
```

```python
theta0_det,theta1_det=linear(0.0005)
```

```python
print(theta0_list,theta1_list) # 0.014631902617044666 0.9987579287601395
```
